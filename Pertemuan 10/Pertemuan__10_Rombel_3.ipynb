Dictionary

```python
##Dictionary
#List -> array, mengakses dengan menggunakan index
data_list = ["Sylus", "Caleb", "Zayne"]
print(data_list[0])

#dictionary (dict)-> associative array
#identifier -> key
data_dict = {
    "key": "value",
    "Sy": "Sylus",
    "Cl": "Caleb",
    "Zy": "Zayne",
    "nmbr": 100 ,
    "list": data_list
}
print(data_dict)
print(data_dict["Zy"])
print(data_dict["nmbr"])
print(data_dict["list"])
print(data_dict)
```

Operasi Dictionary


```python
##--Operasi Dictionary --

data_dict = {
    "Lus": "Sylus leader",
    "Leb": "Caleb kolonel",
    "Yel": "Rafayel seagod"
}
#Panjang dictionary
LENDICT = len(data_dict)
print(f"Panjang dictionary : {LENDICT}") #/(dict_len)

#Mengecek key exist atau tidak
key = "Lus"
check_key = key in data_dict
print(f"Apakah {key} ada di data_dict: {check_key}")

#Mengakses value (read) dengan get
print(data_dict ["Lus"])
print(data_dict.get("Lus"))
print(data_dict.get ("kis", "key tidak ditemukan"))#mengecek key dengan message

#Mengupdate data
data_dict["Lus"] = "Sylus si ganteng"
print(f"\ndata sebelum  update:", data_dict)
data_dict["Xa"] = "Xavier si raja"
print(f"\ndata sebelum di update:\n", data_dict)

data_dict.update({"Lus": "Sylus the dragon"})
print(f"\ndata setelah diupdate:\n", data_dict)
data_dict.update({"Zyn": "Zayne emang keren"})
#akan ditambah jika tidak ada
print(f"\ndata setelah di update:\n",data_dict)

#Menghapus data
del data_dict ["Zyn"]
print(f"\ndata setelah di hapus:",data_dict)
```

Looping Dictionary

```python
##--Looping Dictionary

teman_teman = {
    "Lus": "Sylus leader",
    "Mgy": "Mingyu SVT",
    "Njm": "Na Jaemin NCT",
    "Jno": "Jeno NCT"
}
#Looping first try (yang keluar hanyalah key)
for teman in teman_teman:
    print("Key:", teman)
```

Latihan/ Tugas

```python
# a. Membuat Tuple
mahasiswa = ("Nabilla Dwi Putri Paraswati", "2505060059", "Teknologi Informasi")
print(mahasiswa)

# b. Mengubah prodi menjadi "Teknologi Informasi dan Elektro"
#karena tuple bersifat immutable maka  buat isi baru
mahasiswa = (mahasiswa [0], mahasiswa[1], "Teknologi Informasi dan Elektro")
print(mahasiswa)

# c. menampilkan NIM dan Prodi
print("\nNIM :", mahasiswa [1])
print("\nProdi :", mahasiswa[2])
```

```python
# a. Membuat dictionary
keranjang = {
    "apel": 5000,
    "pisang": 3000,
    "jeruk": 7000,
    "anggur": 15000
    }
print(f"Keranjang :",keranjang)

# b. Menghitung total harga semua barang (perulangan)
print("\nTotal harga:")
total = 0
for harga in keranjang.values():
  total += harga
  print(f"Total harga",{harga})

# c. Menampilkan setiap barang dan harganya
print("\nHarga tiap barang")
for buah, harga in keranjang.items():
  print(f"Harga {buah}: Rp {harga}")

# d. Menampilkan total belanja
print("\nTotal belanja")
print("Total harga belanja :RP", total)
```
